```python
years = np.arange(2018,2026)



def main_df():
    for year in years:
        df = pd.read_csv(f'../data/Metric/{year}/metric scores.csv')
        df['year'] = year
        if year == 2018:
            df_all = df
        else:
            df_all = pd.concat([df_all, df])

    return df_all.loc[df_all['Unnamed: 0'] == 'Dokuz Eylul University'].drop(columns=['Unnamed: 0'])



```

```python
df = main_df()
```

```python
df = df[['year'] + [col for col in df.columns if col != 'year']]
```

```python
for col in df.columns:
    df[col] = df[col].astype(float)
```

```python
df['year'] = df['year'].astype(int)
```

```python
#save
df.to_excel('../data/result/DEU_Metric.xlsx')
```
