```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../data/THE_v2.xlsx')
```

```python
cities = [
    'Istanbul', 'Ankara', 'Istanbul', 'Istanbul', 'Ankara', 'Istanbul',
    'Ankara', 'Istanbul', 'Erzurum', 'Istanbul', 'Konya', 'Antalya',
    'Izmir', 'Kayseri', 'Elazığ', 'Ankara', 'Kocaeli', 'Istanbul',
    'Ankara', 'Bursa', 'Adana', 'Izmir', 'Gaziantep', 'Izmir', 'Trabzon',
    'Kocaeli', 'Istanbul', 'Samsun', 'Sakarya'
]

regions = [
    'Marmara', 'İç Anadolu', 'Marmara', 'Marmara', 'İç Anadolu', 'Marmara',
    'İç Anadolu', 'Marmara', 'Doğu Anadolu', 'Marmara', 'İç Anadolu',
    'Akdeniz', 'Ege', 'İç Anadolu', 'Doğu Anadolu', 'İç Anadolu',
    'Marmara', 'Marmara', 'İç Anadolu', 'Marmara', 'Akdeniz', 'Ege',
    'Güneydoğu Anadolu', 'Ege', 'Karadeniz', 'Marmara', 'Marmara',
    'Karadeniz', 'Marmara'
]



```

```python
df['City'] = cities
df['Region'] = regions
df['City'] = df['City'].str.upper()
df['Region'] = df['Region'].str.upper()
```

```python
df['scores_overall_rank'] = df['scores_overall_rank'] / 10 
```

```python
df_country = df[['name','Region','City']]
```

```python
df_country
```

```python
df.drop(columns=['disabled','nid','url','member_level','record_type',],inplace=True)
```

```python
df
```

```python
df_2 = df[['name','Region','City','scores_overall','scores_citations','scores_research','scores_teaching','scores_international_outlook','scores_industry_income']]


```

```python
melt_score_df = pd.melt(df_2, id_vars=['name','Region','City'], value_vars=['scores_overall','scores_citations','scores_research','scores_teaching','scores_international_outlook','scores_industry_income'], var_name='score_type', value_name='score')
```

```python
df_3 = df[['name','Region','City','stats_number_students','stats_student_staff_ratio','stats_pc_intl_students','stats_female_male_ratio']]
```

```python
df_3['stats_pc_intl_students'] = df_3['stats_pc_intl_students'].str.replace('%','').astype(float)
df_3['stats_female_ratio'] = df_3['stats_female_male_ratio'].str.split(':').str[0].astype(float)
df_3['stats_male_ratio'] = df_3['stats_female_male_ratio'].str.split(':').str[1].astype(float)
df_3.drop(columns='stats_female_male_ratio',inplace=True)
```

```python
df_student_melt = pd.melt(df_3, id_vars=['name','Region','City'], value_vars=['stats_number_students','stats_student_staff_ratio','stats_pc_intl_students','stats_female_ratio','stats_male_ratio'], var_name='student_type', value_name='student-stat')
```

```python
#save to excel

with pd.ExcelWriter('../data/THE_v4_cleaned.xlsx') as writer:
    df_3.to_excel(writer, sheet_name='main')
    melt_score_df.to_excel(writer, sheet_name='scores')
    df_student_melt.to_excel(writer, sheet_name='students')
    
    

```
